```python
import pandas as pd
df_train = pd.read_csv('/content/ml_project1_data.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mne
from mne.viz import circular_layout
from mne_connectivity.viz import plot_connectivity_circle

# Selecionando apenas as colunas numéricas do DataFrame
numeric_cols = df_train.select_dtypes(include=np.number)

# Obtendo a matriz de correlação
corrMatrix = numeric_cols.corr().values

# Obtendo os nomes das features
node_names = numeric_cols.columns.tolist()

# Criando a figura e o objeto PolarAxes
fig = plt.figure(figsize=(15,10 ))
ax = fig.add_subplot(111, polar=True)

# Plotando o gráfico circular de correlação com escala de cores 'RdBu'
plot_connectivity_circle(corrMatrix, node_names, n_lines=100, fig=fig, ax=ax,colormap='RdBu',vmin=-1,vmax=1,facecolor='white',textcolor='black')

# Ajustando a posição dos nomes das regiões
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

# Mostrando o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mne
from mne.viz import circular_layout
from mne_connectivity.viz import plot_connectivity_circle

# Selecionar apenas as colunas numéricas do DataFrame
numeric_cols = df_train.select_dtypes(include=np.number)

# Obter a matriz de correlação
corrMatrix = numeric_cols.corr().values

# Obter os nomes das features
node_names = numeric_cols.columns.tolist()

# Criar a figura e o objeto PolarAxes
fig = plt.figure(figsize=(15,10 ))
ax = fig.add_subplot(111, polar=True)

# Plotar o gráfico circular de correlação com escala de cores 'RdBu'
plot_connectivity_circle(corrMatrix, node_names, n_lines=300, fig=fig, ax=ax,colormap='RdBu',vmin=-1,vmax=1,textcolor='red')

# Ajustar a posição dos nomes das regiões
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

# Mostrar o gráfico
plt.show()
```
