```python
import pandas as pd
df_train = pd.read_csv('/content/ml_project1_data.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mne
from mne.viz import circular_layout
from mne_connectivity.viz import plot_connectivity_circle

# Selecionando apenas as colunas numéricas do DataFrame
numeric_cols = df_train.select_dtypes(include=np.number)

# Obtendo a matriz de correlação
corrMatrix = numeric_cols.corr().values

# Obtendo os nomes das features
node_names = numeric_cols.columns.tolist()

# Criando a figura e o objeto PolarAxes
fig = plt.figure(figsize=(15,10 ))
ax = fig.add_subplot(111, polar=True)

# Plotando o gráfico circular de correlação com escala de cores 'RdBu'
plot_connectivity_circle(corrMatrix, node_names, n_lines=100, fig=fig, ax=ax,colormap='RdBu',vmin=-1,vmax=1,facecolor='white',textcolor='black',linewidth= 1)

# Ajustando a posição dos nomes das regiões
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

# Mostrando o gráfico
plt.show()
```

no codigo ax = fig.add_subplot(111, polar=True) cria uma área dentro da figura onde o gráfico circular de correlação será desenhado.

- Podemos imaginar a figura como uma tela em branco e o add_subplot é como desenhar uma caixa dentro dessa tela. A caixa é o espaço reservado para o gráfico polar.

- Aqui, o add_subplot(111) indica que queremos apenas um subplot na figura (uma única caixa) e (polar=True) significa que queremos que esse subplot seja um gráfico polar, ou seja, um gráfico circular.

- Em seguida, armazenamos essa caixa (subplot) em uma variável chamada ax. Essa variável ax é usada posteriormente para passar como argumento para a função plot_connectivity_circle, que irá desenhar o gráfico circular de correlação dentro dessa caixa específica.

- Resumindo, essa linha cria uma "caixa" dentro da figura onde o gráfico circular de correlação será desenhado.
